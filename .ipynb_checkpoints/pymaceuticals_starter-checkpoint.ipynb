```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merge_df = pd.merge(study_results, mouse_metadata, on="Mouse ID", how="left")

# Display the data table for preview
merge_df
```

```python
# Checking the number of mice.
mouse_count = len(merge_df["Mouse ID"].unique())
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mouse_ID = merge_df.loc[merge_df.duplicated(subset=["Mouse ID", "Timepoint"]),"Mouse ID"].unique()
duplicate_mouse_ID
                                  
```

```python
# Optional: Get all the data for the duplicate mouse ID.
index_df = merge_df.set_index('Mouse ID')
index_df.loc["g989"]

```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = index_df.drop(['g989'])
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
clean_df2 = clean_df.reset_index()
mouse_count_clean = len(clean_df2["Mouse ID"].unique())
mouse_count_clean
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
# Assemble the resulting series into a single summary DataFrame.

tumor = clean_df2.groupby(['Drug Regimen'])

tumor_mean = tumor['Tumor Volume (mm3)'].mean()
tumor_median = tumor['Tumor Volume (mm3)'].median()
tumor_variance = tumor['Tumor Volume (mm3)'].var()
tumor_std = tumor['Tumor Volume (mm3)'].std()
tumor_sem = tumor['Tumor Volume (mm3)'].sem()

tumor_df = pd.DataFrame({"Mean Tumor Volume": tumor_mean,
                         "Median Tumor Volume": tumor_median,
                         "Tumor Volume Variance": tumor_variance,
                         "Tumor Volume Std. Dev.": tumor_std,
                         "Tumor Volume Std. Err.": tumor_sem})
tumor_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
tumor2 = clean_df2.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
tumor2
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin

drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
drugs = clean_df2[clean_df2["Drug Regimen"].isin(drug_list)]
drugs.head()

# Start by getting the last (greatest) timepoint for each mouse

last_timepoint = drugs.groupby("Mouse ID")["Timepoint"].max().reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
final_tumor_volume_df = last_timepoint.merge(drugs, on=["Mouse ID", "Timepoint"], how="left")

#Filter for the four treatment regimens
final_tumor_volume_df = final_tumor_volume_df[final_tumor_volume_df["Drug Regimen"].isin(drug_list)]

# Display the final dataset
print(final_tumor_volume_df[["Mouse ID", "Drug Regimen", "Timepoint", "Tumor Volume (mm3)"]])
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volumes = {}


# Calculate the IQR and quantitatively determine if there are any potential outliers.
for drug in treatments:
    # Get the tumor volume data for the current treatment
    drug_data = final_tumor_volume_df[final_tumor_volume_df["Drug Regimen"] == drug]["Tumor Volume (mm3)"]
    
    # Add to dictionary for later plotting
    tumor_volumes[drug] = drug_data

    # Calculate IQR
    quartiles = np.percentile(drug_data, [25, 75])
    lowerq, upperq = quartiles
    iqr = upperq - lowerq

    # Calculate lower and upper bounds
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)

    # Find potential outliers
    outliers = drug_data[(drug_data < lower_bound) | (drug_data > upper_bound)]
    
    print(f"\n{drug} - Potential Outliers:")
    print(outliers if not outliers.empty else "No outliers detected")

    # Locate the rows which contain mice on each drug and get the tumor volumes


    # add subset


    # Determine outliers using upper and lower bounds
```
